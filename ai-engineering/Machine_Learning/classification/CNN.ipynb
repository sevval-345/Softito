```python
import os
import numpy as np
import cv2

# 1. Klasör yollarını tanımlayalım
paths = [
    "dataset/train/cats",
    "dataset/train/dogs",
    "dataset/test/cats",
    "dataset/test/dogs"
]

# 2. Klasörleri bilgisayarda/Colab'de otomatik açalım
print("📁 Klasör yapıları oluşturuluyor...")
for p in paths:
    os.makedirs(p, exist_ok=True)
print("✅ Klasörler hazır!\n")

# 3. Modelin çalışabilmesi için her klasöre 10'ar tane örnek sahte resim üretelim
print("🖼️ Eğitim için sahte kedi/köpek görselleri üretiliyor...")
for p in paths:
    for i in range(10):
        # 150x150 boyutunda rastgele renkli piksellerden oluşan bir matris (sahte resim)
        rastgele_resim = np.random.randint(0, 256, (150, 150, 3), dtype=np.uint8)
        resim_adi = f"img_{i}.jpg"
        tam_yol = os.path.join(p, resim_adi)
        cv2.imwrite(tam_yol, rastgele_resim)

print("🎉 Altyapı tamamen kuruldu! Sol paneli yenileyerek 'dataset' klasörünü görebilirsin.")
print("🚀 Şimdi ana CNN kodunu içeren hücreyi yeniden çalıştırabilirsin!")
```

```python
import os
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Rescaling, Input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# ==============================================================================
# 1. ADIM: VERI YOLLARI VE HİPERPARAMETRELER
# ==============================================================================
TRAIN_DIR = "dataset/train"
TEST_DIR = "dataset/test"

IMG_HEIGHT = 150
IMG_WIDTH = 150
BATCH_SIZE = 4
EPOCHS = 15

# ==============================================================================
# 2. ADIM: VERI ÖN IŞLEME VE AGRESİF DATA AUGMENTATION
# ==============================================================================
print("🔄 Veri yükleyiciler ve Geliştirilmiş Data Augmentation hazırlanıyor...")

# Modelin ezberini bozmak için görüntü çoğaltma oranlarını biraz daha artırdık
train_datagen = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

test_datagen = ImageDataGenerator()

train_generator = train_datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=True
)

test_generator = test_datagen.flow_from_directory(
    TEST_DIR,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False
)

# ==============================================================================
# 3. ADIM: OVERFITTING ENGELLİ HAFİF CNN MİMARİSİ
# ==============================================================================
print("\n🏗️ Ezber Engelli (Düzeltilmiş) CNN Mimarisi İnşa Ediliyor...")

model = Sequential([
    # Keras 3 Uyarısını Çözen Giriş Katmanı
    Input(shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
    Rescaling(1./255),

    # 1. Evrişim Bloğu: Filtre sayısını 32'den 16'ya düşürdük (Kapasite kısıtlama)
    Conv2D(16, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    # 2. Evrişim Bloğu: Filtre sayısını 64'ten 32'ye düşürdük
    Conv2D(32, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    # Matrisi düzleştirme
    Flatten(),

    # Yorumlama Katmanı: Nöron sayısını 128'den 64'ü düşürerek ezber alanını daralttık
    Dense(64, activation='relu'),

    # Dropout oranını %50'den %60'a çıkararak bağları daha sıkı kopardık
    Dropout(0.6),

    # Çıkış Katmanı
    Dense(1, activation='sigmoid')
])

# Yeni hafifletilmiş modelin özetini yazdırır (Parametre sayısı ciddi oranda düştü)
model.summary()

# ==============================================================================
# 4. ADIM: MODELİN DERLENMESİ
# ==============================================================================
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# ==============================================================================
# 5. ADIM: MODELİN EĞİTİLMESİ
# ==============================================================================
print(f"\n🚀 Düzeltilmiş CNN Eğitimi Başlıyor ({EPOCHS} Epoch)...")

history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=test_generator
)

print("✅ Eğitim tamamlandı!")

# ==============================================================================
# 6. ADIM: BAŞARI VE KAYIP GRAFİKLERİNİN ÇİZİLMESİ
# ==============================================================================
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(EPOCHS)

plt.figure(figsize=(12, 5))

# Doğruluk Grafiği
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Egitim Basarisi')
plt.plot(epochs_range, val_acc, label='Test Basarisi')
plt.title('Egitim ve Test Dogruluk Oranlari')
plt.xlabel('Epoch')
plt.ylabel('Dogruluk')
plt.legend(loc='lower right')

# Kayıp Grafiği
plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Egitim Kaybi')
plt.plot(epochs_range, val_loss, label='Test Kaybi')
plt.title('Egitim ve Test Kayip Degerleri')
plt.xlabel('Epoch')
plt.ylabel('Kayip (Loss)')
plt.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

```python
# Modelin ulaştığı son başarı oranlarını sayısal olarak görelim
son_egitim_dogrulugu = history.history['accuracy'][-1] * 100
son_test_dogrulugu = history.history['val_accuracy'][-1] * 100

print("="*45)
print(f"🎯 Son Eğitim Doğruluğu (Train Accuracy): %{son_egitim_dogrulugu:.2f}")
print(f"🧪 Son Doğrulama Doğruluğu (Test Accuracy): %{son_test_dogrulugu:.2f}")
print("="*45)
```
