### 🔸 Soru 10:
`while` döngüsü ile basamak sayısı bulma: Verilen bir sayının kaç basamaklı olduğunu, sayıyı 10'a bölerek bulun (string dönüşümü kullanmadan).

```python
# Cevabınızı buraya yazın:
# Örnek: 4752 -> 4 basamaklı

sayi = 4752
basamak_sayisi = 0

# Eğer sayı 0 ise özel olarak ele alınmalı (çünkü 0, 1 basamaklıdır)
if sayi == 0:
    basamak_sayisi = 1
else:
    # Sayı negatifse pozitife çeviriyoruz
    gecici_sayi = abs(sayi)

    while gecici_sayi > 0:
        gecici_sayi //= 10  # Sayıyı 10'a tam bölerek bir basamak eksilt
        basamak_sayisi += 1 # Her bölmede sayacı 1 artır

print(f"Basamak sayısı: {basamak_sayisi}")
```

### 🔸 Soru 12:
Bir öğrenci listesi var. Her öğrenci `(isim, not)` tuple'ı olarak saklanıyor.
- Notu 70'in üzerinde olanları filtreleyin
- En yüksek notu alan öğrenciyi bulun
- Notların ortalamasını hesaplayın

```python
# Cevabınızı buraya yazın:
ogrenciler = [
    ("Ahmet", 85),
    ("Büşra", 62),
    ("Can", 91),
    ("Deniz", 74),
    ("Elif", 58),
    ("Furkan", 88)
]

# --- 1. Filtreleme (70'ten büyük olanlar) ---
basarili_ogrenciler = []
for ogrenci in ogrenciler:
    if ogrenci[1] > 70:
        basarili_ogrenciler.append(ogrenci)

# --- 2. En Yüksek Notu Bulma ---
en_yuksek_ogrenci = ogrenciler[0] # İlk öğrenciyi şampiyon varsayalım
for ogrenci in ogrenciler:
    if ogrenci[1] > en_yuksek_ogrenci[1]:
        en_yuksek_ogrenci = ogrenci

# --- 3. Ortalama Hesaplama ---
toplam_puan = 0
for ogrenci in ogrenciler:
    toplam_puan += ogrenci[1]

ortalama = toplam_puan / len(ogrenciler)

# --- SONUÇLARI YAZDIRMA ---
print("1- Başarılı Öğrenciler (70+):", basarili_ogrenciler)
print("2- Sınıf Birincisi:", en_yuksek_ogrenci[0], "Notu:", en_yuksek_ogrenci[1])
print("3- Sınıf Ortalaması:", ortalama)
```

### 🔸 Soru 15:
Özyinelemeli (recursive) bir fonksiyon yazın: `faktoriyel(n)` - n! hesaplasın.
- 0! = 1
- n! = n × (n-1)!

5!, 10! değerlerini hesaplayın.

```python
# Cevabınızı buraya yazın:
def faktoriyel(n):
    # Temel durum (Döngünün duracağı yer)
    if n == 0:
        return 1
    # Özyineleme adımı
    else:
        return n * faktoriyel(n - 1)

# Testler
print(f"5! = {faktoriyel(5)}")   # 120
print(f"10! = {faktoriyel(10)}") # 3,628,800
```

### 🔸 Soru 16:
`lambda` fonksiyonu kullanarak:
- Bir sayının karesini hesaplayan lambda
- İki sayının büyüğünü döndüren lambda
- `sorted()` ile bir isim listesini son harfe göre sıralayan lambda

```python
# Cevabınızı buraya yazın:
isimler = ["Zeynep", "Ali", "Mahmut", "Su", "Berk"]

# --- 1. KARE ALMA ---
# x girer, x*x çıkar
kare_al = lambda x: x * x
print(f"5'in karesi: {kare_al(5)}")


# --- 2. BÜYÜK SAYIYI BULMA ---
# a ve b girer, hangisi büyükse o seçilir
buyuk_olani_bul = lambda a, b: a if a > b else b
print(f"10 mu, 25 mi daha büyük?: {buyuk_olani_bul(10, 25)}")


# --- 3. SON HARFE GÖRE SIRALAMA ---


# sorted() fonksiyonuna 'key' vererek son harfe (-1) bakmasını söylüyoruz
sirali_liste = sorted(isimler, key=lambda x: x[-1])

print(f"Son harfe göre sıralama: {sirali_liste}")
```

### 🔸 Soru 17:
Bir alışveriş listesini `alisveris.txt` dosyasına yazın (her ürün ayrı satırda). Sonra dosyayı okuyarak ürünleri numaralı şekilde ekrana yazdırın.

```python
# Cevabınızı buraya yazın:
urunler = ["ekmek", "süt", "yumurta", "peynir", "zeytin"]


# 1. AŞAMA: Dosyaya Yazma
# 'w' (write) modu: Dosya yoksa oluşturur, varsa içini silip baştan yazar.
with open("alisveris.txt", "w", encoding="utf-8") as dosya:
    for urun in urunler:
        dosya.write(urun + "\n") # Her üründen sonra alt satıra geçmesi için \n ekledik

print("Dosya başarıyla oluşturuldu!\n")

# 2. AŞAMA: Dosyadan Okuma ve Numaralandırma
print("--- ALIŞVERİŞ LİSTESİ ---")
# 'r' (read) modu: Dosyayı okumak için açar.
with open("alisveris.txt", "r", encoding="utf-8") as dosya:
    # Her bir satırı (urun) al ve enumerate ile numaralandır
    for sira, urun in enumerate(dosya, 1):
        # strip() fonksiyonu satır sonundaki görünmez boşlukları/alt satır karakterlerini siler
        print(f"{sira}. {urun.strip()}")
```

### 🔸 Soru 18 (Bütünleştirici):
Basit bir **öğrenci not sistemi** yazın:

- `ogrenciler` adında boş bir sözlük oluşturun (anahtar: isim, değer: notlar listesi)
- `ogrenci_ekle(isim)`: Yeni öğrenci ekler
- `not_ekle(isim, not_)`: Öğrenciye not ekler. Öğrenci yoksa hata mesajı verir.
- `ortalama_hesapla(isim)`: Öğrencinin not ortalamasını döndürür
- `sinif_raporu()`: Tüm öğrencileri ve ortalamalarını yazdırır, sınıf ortalamasını gösterir

En az 3 öğrenci ve her birine 3 not girerek test edin.

```python
# Cevabınızı buraya yazın:
# Başlangıçta boş bir sözlük (Dictionary)
ogrenciler = {}

def ogrenci_ekle(isim):
    # Eğer isim sözlükte yoksa, ona boş bir not listesi açalım
    if isim not in ogrenciler:
        ogrenciler[isim] = []
        print(f"Sisteme Eklendi: {isim}")
    else:
        print(f"Uyarı: {isim} zaten kayıtlı!")

def not_ekle(isim, not_):
    # Önce bu isimde bir öğrenci var mı kontrol edelim
    if isim in ogrenciler:
        ogrenciler[isim].append(not_)
    else:
        print(f"Hata: {isim} bulunamadı, not eklenemedi.")

def ortalama_hesapla(isim):
    notlar = ogrenciler[isim]
    if not notlar: # Hiç notu yoksa hata vermemesi için 0 döndürelim
        return 0
    return sum(notlar) / len(notlar)

def sinif_raporu():
    print("\n" + "="*30)
    print("      SINIF RAPORU")
    print("="*30)

    toplam_ortalamalar = 0

    # Her bir öğrenciyi numaralandırarak yazdıralım
    for sira, isim in enumerate(ogrenciler, 1):
        ort = ortalama_hesapla(isim)
        toplam_ortalamalar += ort
        print(f"{sira}. {isim:8} | Notlar: {ogrenciler[isim]} | Ort: {ort:.2f}")

    # Sınıf genel ortalaması (öğrenci sayısı 0 değilse hesapla)
    if len(ogrenciler) > 0:
        genel_ort = toplam_ortalamalar / len(ogrenciler)
        print("-"*30)
        print(f"Sınıf Genel Ortalaması: {genel_ort:.2f}")
    print("="*30)

# --- Test Kodu ---
ogrenci_ekle("Ahmet")
ogrenci_ekle("Büşra")
ogrenci_ekle("Can")

not_ekle("Ahmet", 85)
not_ekle("Ahmet", 90)
not_ekle("Ahmet", 78)

not_ekle("Büşra", 92)
not_ekle("Büşra", 88)
not_ekle("Büşra", 95)

not_ekle("Can", 70)
not_ekle("Can", 65)
not_ekle("Can", 73)

# Sonuçları ekrana bas
sinif_raporu()
```
