1. Temel Hata Yönetimi (try - except)


```python
# Kullanıcıdan sayı alırken oluşabilecek hataları yakalama
try:
    sayi = int(input("Lütfen bir sayı girin: "))
    sonuc = 100 / sayi
    print(f"Sonuç: {sonuc}")

except ZeroDivisionError:
    # Kullanıcı 0 girerse çalışacak blok
    print("Hata: Bir sayı sıfıra bölünemez!")

except ValueError:
    # Kullanıcı sayı yerine harf girerse çalışacak blok
    print("Hata: Lütfen sadece sayısal bir değer girin!")
```

2. Gelişmiş Bloklar (else ve finally)


```python
try:
    print("Veritabanına bağlantı açılıyor...")
    dosya_sirasi = int(input("Okumak istediğiniz satır numarası: "))
    # Varsayalım ki sadece 10 satır var, kullanıcı 50 girerse hata verecektir
    if dosya_sirasi > 10:
        raise IndexError  # Yapay olarak hata fırlatıyoruz

except IndexError:
    print("Hata: Geçersiz satır numarası girdiniz!")

else:
    print("Tebrikler! İşlem başarıyla tamamlandı, veri okundu.")

finally:
    # Bu blok temizlik işleri için çok önemlidir (Dosya kapatma, bağlantı kesme vb.)
    print("Veritabanı bağlantısı güvenli bir şekilde kapatıldı.")
```

3. Python'da En Sık Karşılaşılan Hata Türleri (Örnekler)

```python
# 1. ValueError (Değer Hatası)
# int() fonksiyonu "yirmi" metnini sayıya çeviremez.
try:
    sayi = int("yirmi")
except ValueError as e:
    print(f"ValueError Yakalandı: {e}")

# 2. NameError (Tanımsız Değişken Hatası)
# 'mesaj' adında bir değişken önceden tanımlanmadığı için hata verir.
try:
    print(mesaj)
except NameError as e:
    print(f"NameError Yakalandı: {e}")

# 3. TypeError (Tip/Tür Hatası)
# Bir metin (str) ile bir sayı (int) toplanamaz.
try:
    sonuc = "Yaşınız: " + 25
except TypeError as e:
    print(f"TypeError Yakalandı: {e}")

# 4. IndexError (Liste Sınırı Hatası)
# Listenin sadece 3 elemanı var (0, 1, 2). 5. elemana ulaşmaya çalışmak hata verir.
try:
    liste = [10, 20, 30]
    print(liste[5])
except IndexError as e:
    print(f"IndexError Yakalandı: {e}")

# 5. KeyError (Sözlük Anahtar Hatası)
# 'sozluk' içinde 'eposta' adında bir anahtar (key) yok.
try:
    sozluk = {"isim": "Ahmet", "yas": 30}
    print(sozluk["eposta"])
except KeyError as e:
    print(f"KeyError Yakalandı: {e}")
```

4.Dosya ve Sistem Hataları

```python
# 1. FileNotFoundError (Dosya Bulunamadı Hatası)
# Bilgisayarda var olmayan bir dosyayı açmaya (okumaya) çalıştığımızda oluşur.
try:
    with open("olmayan_dosya.txt", "r") as dosya:
        icerik = dosya.read()
except FileNotFoundError as e:
    print(f"FileNotFoundError Yakalandı: {e}")
    print("İpucu: Dosya yolunu veya dosyanın gerçekten var olduğunu kontrol edin.\n")


# 2. PermissionError (Erişim/İzin Hatası)
# Yetkinizin olmadığı bir dosyayı değiştirmeye veya bir klasörü dosya gibi açmaya çalıştığınızda oluşur.
try:
    # Windows veya Linux'ta sistem korumalı bir klasörü açmaya çalışmak gibi
    with open("/etc/shadow", "r") as dosya:
        print(dosya.read())
except PermissionError as e:
    print(f"PermissionError Yakalandı: {e}")
    print("İpucu: Bu işlemi yapmak için yönetici (admin) izniniz yok.\n")
```

5.Matematiksel ve Sayısal Hatalar

```python
# 3. OverflowError (Taşma Hatası)
# Bir matematiksel işlemin sonucu, Python'ın bellekte o sayı türü için ayırabileceği sınırdan büyük olduğunda oluşur.
import math
try:
    # Sayısal olarak hesaplanamayacak kadar devasa bir üst alma işlemi
    sonuc = math.exp(1000)
except OverflowError as e:
    print(f"OverflowError Yakalandı: {e}")
    print("İpucu: Hesaplanan sayı bilgisayarın kapasitesi için çok büyük.\n")
```

6. Kod Yapısı ve Modül Hataları

```python
# 4. ModuleNotFoundError (Modül Bulunamadı Hatası)
# Bilgisayarınızda veya Colab ortamında yüklü olmayan bir kütüphaneyi "import" etmeye çalışırsanız oluşur.
try:
    import yapay_zeka_kutuphanesi  # Böyle bir kütüphane yok
except ModuleNotFoundError as e:
    print(f"ModuleNotFoundError Yakalandı: {e}")
    print("İpucu: Kodu çalıştırmadan önce 'pip install kutuphane_adi' yapmanız gerekebilir.\n")


# 5. AttributeError (Öznitelik Hatası)
# Bir veri tipine veya nesneye, onda var olmayan bir fonksiyonu ya da özelliği uygulaya çalıştığınızda oluşur.
try:
    sayi = 10
    # Sayılar (int) metinler gibi büyütülemez, .upper() fonksiyonu sayılarda yoktur!
    sayi.upper()
except AttributeError as e:
    print(f"AttributeError Yakalandı: {e}")
    print("İpucu: Kullandığınız veri tipinin o metoda/fonksiyona sahip olduğundan emin olun.\n")
```

7. Bellek ve Döngü Hataları

```python
# 6. RecursionError (Sonsuz Özyineleme Hatası)
# Bir fonksiyon kendi kendini sonsuz defa çağırdığında (çıkış koşulu olmadan) oluşur.
def sonsuz_fonksiyon():
    return sonsuz_fonksiyon()

try:
    sonsuz_fonksiyon()
except RecursionError as e:
    print(f"RecursionError Yakalandı: {e}")
    print("İpucu: Fonksiyon kendi kendini çok fazla çağırdı ve Python'ın güvenlik sınırını (stack) aştı.\n")
```
