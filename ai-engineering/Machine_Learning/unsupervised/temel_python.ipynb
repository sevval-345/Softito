### 🔸 Soru 5:
`cumle = "  python programlama dili çok güçlüdür  "` değişkeni için:
- Baştaki ve sondaki boşlukları temizleyin (`.strip()`)
- İlk harfini büyük yapın (`.capitalize()`)
- Kaç kelimeden oluştuğunu bulun (`.split()` ve `len()` kullanın)
- "güçlü" kelimesinin cümlede olup olmadığını kontrol edin (`in` operatörü)

```python
# Cevabınızı buraya yazın:

cumle = " python programlama dili çok güçlüdür "

# 1. Boşlukları temizle
temiz_cumle = cumle.strip()

# 2. İlk harfi büyük yap
print(temiz_cumle.capitalize())

# 3. Kelime sayısını bul
kelime_sayisi = len(temiz_cumle.split())
print(f"Kelime sayısı: {kelime_sayisi}")

# 4. "güçlü" kelimesini kontrol et
print("güçlü" in temiz_cumle)
```

### 🔸 Soru 9:
`for` döngüsü ile 1'den 10'a kadar olan sayıların toplamını hesaplayın.

```python
# Cevabınızı buraya yazın:
# Başlangıçta toplamı sıfır olarak tanımlıyoruz
toplam = 0

# range(1, 11) komutu 1'den başlar, 11'e kadar (11 hariç) sayılar üretir
for sayi in range(1, 11):
    toplam += sayi  # Her adımda sayıyı toplam değişkenine ekler

print(f"1'den 10'a kadar olan sayıların toplamı: {toplam}")
```

### 🔸 Soru 10:
`while` döngüsü ile kullanıcının girdiği sayıya (n=5) kadar Fibonacci dizisini yazdırın.

Örnek çıktı: `0, 1, 1, 2, 3`

```python
# Cevabınızı buraya yazın:
# Kullanıcının girdiği sınır değeri
n = 5

# Fibonacci dizisinin ilk iki terimi
a, b = 0, 1
sayac = 0

print(f"{n} terimli Fibonacci dizisi:")

# Sayaç n değerine ulaşana kadar devam eder
while sayac < n:
    # Sayıyı yazdır (virgül ile yan yana gelmesi için end parametresini kullanıyoruz)
    if sayac == n - 1:
        print(a)
    else:
        print(a, end=", ")

    # Yeni sayıları hesapla (a bir sonraki sayı olur, b ise toplamları)
    yeni_sayi = a + b
    a = b
    b = yeni_sayi

    # Sayacı bir artır
    sayac += 1
```

### 🔸 Soru 11 (İç İçe Döngü):
5x5 çarpım tablosunu yazdırın. Her satır şu formatta olsun:

`1x1=1  1x2=2  1x3=3 ...`

```python
# Cevabınızı buraya yazın:

# Dış döngü: 1'den 5'e kadar olan satırlar için
for i in range(1, 6):
    # İç döngü: Her satırdaki 1'den 5'e kadar olan çarpımlar için
    for j in range(1, 6):
        # Çarpımı hesapla ve yan yana yazdır
        # end=" " kullanarak bir sonraki print'in aynı satıra devam etmesini sağlarız
        print(f"{i}x{j}={i*j}", end=" ")

    # Bir satır bittiğinde bir alt satıra geçmek için boş bir print atıyoruz
    print()
```

### 🔸 Soru 12:
Aşağıdaki listede:
- En büyük ve en küçük elemanı bulun
- Listeyi sıralayın
- 7 sayısının listede kaç kez geçtiğini bulun
- Listeyi tersine çevirin

```python
# Cevabınızı buraya yazın:
sayilar = [3, 7, 1, 9, 7, 4, 7, 2, 8]

# 1. En büyük ve en küçük elemanı bulun
en_buyuk = max(sayilar)
en_kucuk = min(sayilar)
print(f"En büyük sayı: {en_buyuk}")
print(f"En küçük sayı: {en_kucuk}")

# 2. 7 sayısının listede kaç kez geçtiğini bulun
kac_tane_yedi = sayilar.count(7)
print(f"7 sayısı listede {kac_tane_yedi} kez geçiyor.")

# 3. Listeyi sıralayın (Küçükten büyüğe)
sayilar.sort()
print(f"Sıralanmış liste: {sayilar}")

# 4. Listeyi tersine çevirin (Büyükten küçüğe veya mevcut sırayı ters çevirme)
sayilar.reverse()
print(f"Ters çevrilmiş liste: {sayilar}")
```

### 🔸 Soru 13 (List Comprehension):
List comprehension kullanarak:
- 1-20 arasındaki çift sayıların listesini oluşturun
- Bir kelime listesindeki 4'ten uzun kelimeleri filtreleyin

```python
# Cevabınızı buraya yazın:
# 1-20 arası çift sayılar
ciftler = [x for x in range(1, 21) if x % 2 == 0]
print(ciftler)

# 4'ten uzun kelimeler
kelimeler = ["at", "kaplan", "it", "fil", "kedi", "aslan", "köpek"]
uzunlar = [k for k in kelimeler if len(k) > 4]
print(uzunlar)
```

### 🔸 Soru 15:
Bir sayının asal olup olmadığını kontrol eden `asal_mi(n)` fonksiyonu yazın.
- 1'den 20'ye kadar olan asal sayıları bu fonksiyonu kullanarak yazdırın.

```python
# Cevabınızı buraya yazın:
def asal_mi(n):
    if n < 2:
      return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0: return False
    return True

# 1'den 20'ye kadar olan asalları yazdır
for sayi in range(1, 21):
    if asal_mi(sayi):
        print(sayi, end=" ")
```

### 🔸 Soru 16:
Bir listenin ortalamasını, en büyük ve en küçük elemanını döndüren `liste_istatistik(liste)` fonksiyonu yazın.
Fonksiyon 3 değer döndürsün: `(ortalama, minimum, maksimum)`

```python
# Cevabınızı buraya yazın:
def liste_istatistik(liste):
    ortalama = sum(liste) / len(liste)
    minimum = min(liste)
    maksimum = max(liste)
    return ortalama, minimum, maksimum

test = [4, 8, 15, 16, 23, 42]
ort, mini, maksi = liste_istatistik(test)

print(f"Ortalama: {ort}, Min: {mini}, Max: {maksi}")
```

### 🔸 Soru 17:
Kullanıcıdan bir sayı alındığını varsayın (`girdi = "abc"`).
Bu değeri int'e çevirmeye çalışırken `ValueError` hatası oluşabilir.
try-except ile bunu yakalayan ve anlamlı bir mesaj veren kod yazın.

```python
# Cevabınızı buraya yazın:
girdi = "abc"

try:
    # Sayıya çevirme işlemini deniyoruz
    sayi = int(girdi)
    print(f"Girdiğiniz sayı: {sayi}")
except ValueError:
    # Eğer çevirme sırasında hata (ValueError) oluşursa burası çalışır
    print("Hata: Girdiğiniz değer bir sayı değil!")
```
